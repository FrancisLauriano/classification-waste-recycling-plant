## üß† Estrat√©gia de Treinamento Utilizada

### ‚úÖ Modelo Proposto: **ResNet50 pr√©-treinada no ImageNet**

A arquitetura **ResNet50** (He et al., 2016) foi escolhida por seu alto desempenho em tarefas de **vis√£o computacional profunda**, especialmente quando combinada com **Transfer Learning**, t√©cnica que permite utilizar o conhecimento aprendido em grandes datasets (como o ImageNet) e aplic√°-lo em problemas espec√≠ficos (Pan & Yang, 2010).

**ResNet50** √© uma arquitetura de redes neurais profundas introduzida por **He et al. (2016)** no artigo *"Deep Residual Learning for Image Recognition"*. Sua principal inova√ß√£o √© o uso de **conex√µes residuais** (atalhos ou skip connections) que permitem treinar redes muito profundas (com dezenas ou centenas de camadas) sem sofrer com o problema de degrada√ß√£o da acur√°cia.

**üìå Principais caracter√≠sticas:**
- 50 camadas profundas;
- Aprendizagem residual (f(x) + x);
- Ampla ado√ß√£o como backbone em aplica√ß√µes com Transfer Learning;
- Excelente equil√≠brio entre profundidade, desempenho e custo computacional.

## üß™ Pipeline Detalhado

### üîπ 1. Pr√©-processamento

- Leitura estruturada dos caminhos e r√≥tulos das imagens, com extra√ß√£o da classe a partir dos diret√≥rios.
- Visualiza√ß√£o da distribui√ß√£o de classes para an√°lise do desbalanceamento.
- **Undersampling seletivo** da classe majorit√°ria `plastic bottles`, reduzindo de **6.070 para 1.070 imagens** (aproximadamente 1/6), enquanto todas as demais classes foram mantidas intactas.
  
üìâ **Essa decis√£o foi tomada com base nos seguintes fatores:**
- A classe `plastic bottles` representava mais de **68%** do total de amostras, o que poderia **enviesar fortemente o modelo** em favor dessa classe, prejudicando o desempenho nas classes minorit√°rias;
- A redu√ß√£o para 1.070 imagens **preserva varia√ß√£o suficiente** dentro da classe majorit√°ria para evitar perda de representatividade;
- O uso de **Data Augmentation**, **Transfer Learning (ResNet50)** e **Valida√ß√£o Cruzada** ajuda a **compensar a diminui√ß√£o do volume de dados**, favorecendo a generaliza√ß√£o;
- Esta t√©cnica segue as melhores pr√°ticas para lidar com dados desbalanceados, conforme discutido por **He & Garcia (2009)**.
- Essa abordagem evita que o modelo se torne enviesado para a classe majorit√°ria, melhorando a equidade na classifica√ß√£o (**He & Garcia, 2009**).

### üîπ 2. Aumenta√ß√£o de Dados (Data Augmentation)

A t√©cnica de **Data Augmentation** √© usada para **aumentar artificialmente a diversidade dos dados de treino**, o que melhora a capacidade de generaliza√ß√£o do modelo (Shorten & Khoshgoftaar, 2019).

‚öôÔ∏è **Configura√ß√µes utilizadas:**

| Par√¢metro | Justificativa |
|-----------|---------------|
| `rescale=1./255` | Normaliza os pixels (0 a 1), o que ajuda na converg√™ncia durante o treinamento. |
| `rotation_range=30` | Permite rota√ß√£o aleat√≥ria de at√© 30¬∞, simulando √¢ngulos de captura variados. |
| `width_shift_range=0.3` | Translada a imagem horizontalmente em at√© 30% da largura. |
| `height_shift_range=0.3` | 	Translada verticalmente at√© 30% da altura, √∫til para simular deslocamentos reais. |
| `shear_range=0.3` | Inclina√ß√£o das imagens, √∫til para lidar com distor√ß√µes de perspectiva. |
| `zoom_range=0.3` | Amplia ou reduz em at√© 30%, simulando dist√¢ncias de captura. |
| `horizontal_flip=True` | 	Espelha horizontalmente ‚Äî comum em imagens sim√©tricas. |
| `vertical_flip=True` | Espelha verticalmente ‚Äî √∫til para objetos capturados de diferentes √¢ngulos. |
| `brightness_range=[0.7, 1.3]` | Altera o brilho da imagem, imitando diferentes condi√ß√µes de ilumina√ß√£o. |
| `channel_shift_range=30.0` | Modifica os canais de cor, ajudando a lidar com varia√ß√µes em ilumina√ß√£o e c√¢meras. |
| `fill_mode='nearest` | Preenche √°reas criadas pelas transforma√ß√µes com os pixels vizinhos mais pr√≥ximos. |

### üîπ 3. Arquitetura do Modelo

‚úÖ **Backbone:**
```plaintext
ResNet50(include_top=False, weights='imagenet')
```
- Carrega uma **rede convolucional pr√©-treinada** no ImageNet (Transfer Learning);
- `include_top=False` remove as camadas de classifica√ß√£o originais para adicionar novas adaptadas ao problema.


‚úÖ **Camadas adicionadas:**
| Camada | Objetivo |
|-----------|---------------|
| `GlobalAveragePooling2D()` | Reduz a dimensionalidade ap√≥s convolu√ß√µes, preservando os recursos mais importantes. |
| `Dropout(0.5)` | Desativa aleatoriamente 50% dos neur√¥nios, reduzindo o overfitting. |
| `Dense(128, activation='relu')` | Camada totalmente conectada para aprendizado de representa√ß√µes complexas. |
| `BatchNormalization()` | 	Normaliza os valores da camada anterior, acelerando o treinamento e melhorando a estabilidade. |
| `Dropout(0.5)` | Segunda camada de regulariza√ß√£o. |
| `Dense(6, activation='softmax')` | Camada de sa√≠da com 6 neur√¥nios (uma para cada classe) e softmax para classifica√ß√£o multiclasse. |


> ‚ÑπÔ∏è A arquitetura foi projetada para reduzir overfitting e melhorar a generaliza√ß√£o, respeitando o equil√≠brio entre capacidade e regulariza√ß√£o.


### üîπ 4. Treinamento com Valida√ß√£o Cruzada K-Fold

O uso de **K-Fold Cross-Validation** (Kohavi, 1995) visa garantir uma **avalia√ß√£o mais robusta** e confi√°vel, particionando os dados em **5 folds**:
| Componente | Justificativa |
|-----------|---------------|
| `KFold(n_splits=5` | Permite que o modelo seja treinado e validado em diferentes subconjuntos dos dados. |
| `shuffle=True` | Embaralha os dados para garantir que os folds sejam aleat√≥rios e bem distribu√≠dos. |
| `random_state=42` | Garante reprodutibilidade. |
| `EarlyStopping(patience=5)` | Evita overfitting interrompendo o treinamento quando n√£o h√° melhora por 5 √©pocas consecutivas. |
| `class_weight=compute_class_weight(...)` | Compensa o desbalanceamento atribuindo maior peso a classes minorit√°rias. |
| `optimizer=Adam(learning_rate=1e-5)` | 	Otimizador est√°vel e adaptativo (Kingma & Ba, 2015). |

## üìà Avalia√ß√£o

Durante o treinamento de cada fold do K-Fold Cross Validation, foram calculadas as seguintes m√©tricas para garantir uma avalia√ß√£o robusta e completa do modelo:

| M√©trica          | O que √©           | Por que √© usada                               |
|--------------------|---------------------------------------------------------------|------------------------------------------------|
| `Accuracy`           | Propor√ß√£o total de classifica√ß√µes corretas em rela√ß√£o ao total de amostras.                                                    | √â uma m√©trica simples e intuitiva para medir desempenho geral. | `Precision (macro)`  | M√©dia da precis√£o por classe, ignorando a propor√ß√£o de amostras de cada classe. | `Recall (macro)`     | M√©dia da sensibilidade (ou taxa de verdadeiros positivos) por classe. Recall = TP / (TP + FN).                                                    | Mede a capacidade do modelo de encontrar todos os exemplos positivos de cada classe. | `F1-Score (macro)`   | M√©dia harm√¥nica entre Precision e Recall, calculada para cada classe e depois promediada (macro).                                                     | Fornece um equil√≠brio entre precis√£o e cobertura, sendo √∫til especialmente em conjuntos de dados desequilibrados, onde se quer evitar tanto falsos positivos quanto falsos negativos.                  |
| `AUC-ROC (macro)`    | √Årea sob a curva ROC, que mede a separabilidade entre classes. No modo macro, √© a m√©dia entre as curvas de cada classe.                                                  | Indica a capacidade do modelo em distinguir entre as classes. Quanto maior o AUC, melhor a distin√ß√£o. √â especialmente √∫til em classifica√ß√£o multiclasse e desbalanceada. [Fawcett, 2006]        |
| `Cohen's Kappa`      | M√©trica de concord√¢ncia que corrige a acur√°cia esperada ao acaso. Valor entre -1 e 1, sendo 1 concord√¢ncia perfeita.           | √ötil para avaliar a qualidade da classifica√ß√£o mesmo em dados desbalanceados, onde a acur√°cia sozinha pode ser enganosa. [Cohen, 1960]|

## üì¶ Imports necess√°rios

````python
import os
import datetime
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import label_binarize
from tensorflow.keras.callbacks import TensorBoard
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import GlobalAveragePooling2D, Dropout, Dense
from tensorflow.keras.metrics import Precision, Recall
from sklearn.model_selection import KFold
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    roc_auc_score,
    cohen_kappa_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
    precision_score,
    recall_score,
    accuracy_score
)
import pandas as pd
import json
from IPython.display import display, HTML
````

## üìÅ Defini√ß√£o dos caminhos dos dados

````python
train_dir = '../data/Warp-C/train_crops' # se estiver trabalhando com pasta local no vscode
# train_dir = '/content/drive/MyDrive/data/Warp-C/train_crops' # se estiver trabalhando com pasta do google drive no google colab

test_dir = '../data/Warp-C/test_crops' # se estiver trabalhando com pasta local no vscode
# test_dir = '/content/drive/MyDrive/data/Warp-C/test_crops' # se estiver trabalhando com pasta do google drive no google colab
````

## üîÄ Criar dataframe com caminho da imagem e r√≥tulo

````python
# Criar dataframe com caminho da imagem e r√≥tulo:

image_paths = []
labels = []

for class_dir in os.listdir(train_dir):
    class_path = os.path.join(train_dir, class_dir)
    for root, _, files in os.walk(class_path):
        for fname in files:
            if fname.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.webp')):  # apenas imagens
                image_paths.append(os.path.join(root, fname))
                labels.append(class_dir)

df = pd.DataFrame({'filename': image_paths, 'label': labels})
````

## üîé An√°lise Explorat√≥ria dos dados

````python
# An√°lise Explorat√≥ria dos dados:

# Gr√°fico de contagem com valores ao lado das barras (dataset completo)
plt.figure(figsize=(12, 6))
ax = sns.countplot(y=df['label'], palette='viridis')
plt.title('Distribui√ß√£o das classes')
plt.xlabel('Contagem')
plt.ylabel('Classe')

# Adicionar valores ao lado de cada barra
for container in ax.containers:
    ax.bar_label(container, label_type='edge', padding=3)


# Calcular e exibir o somat√≥rio total
total = df['label'].count()
plt.text(0.95, 0.05, f'Total: {total}', transform=plt.gca().transAxes,
         fontsize=12, color='black', ha='right', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
````

## üîª Undersampling Seletivo

````python
# Subamostragem da classe majorit√°ria (undersampling):

# Reduzir Plastic bottles para 2000 amostras
df_plastic = df[df['label'] == 'bottle'].sample(n=1071, random_state=42)

# Manter todas as outras classes como est√£o
df_others = df[df['label'] != 'bottle']


# Concatenar e embaralhar
df = pd.concat([df_plastic, df_others], axis=0).sample(frac=1, random_state=42).reset_index(drop=True)

# Novo gr√°fico de contagem com destaque visual
plt.figure(figsize=(12, 6))
ax = sns.countplot(y=df['label'], palette='viridis')
plt.title('Distribui√ß√£o das classes ap√≥s subamostragem (Undersampling)')
plt.xlabel('Contagem')
plt.ylabel('Classe')


# Adicionar r√≥tulos com as quantidades ao lado das barras
for container in ax.containers:
    ax.bar_label(container, label_type='edge', padding=3)

# Calcular e exibir o somat√≥rio total
total = df['label'].count()
plt.text(0.95, 0.05, f'Total: {total}', transform=plt.gca().transAxes,
         fontsize=12, color='black', ha='right', va='bottom', fontweight='bold')


plt.tight_layout()
plt.show()
````

## üß™ Data Augmentation

````python
# Data Augmentation:
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.3,
    zoom_range=0.3,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.7, 1.3],
    channel_shift_range=30.0,
    fill_mode='nearest'
)

validation_datagen = ImageDataGenerator(rescale=1./255)
````

## üß† Defini√ß√£o do modelo com ResNet50

````python
# Defini√ß√£o do modelo com ResNet50:

resnet_model = Sequential([
    ResNet50(include_top=False, input_shape=(224, 224, 3), weights='imagenet'),
    GlobalAveragePooling2D(),
    Dropout(0.5),
    Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    tf.keras.layers.BatchNormalization(),
    Dropout(0.5),
    Dense(6, activation='softmax', kernel_regularizer=tf.keras.regularizers.l2(0.01))
])


resnet_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5, clipnorm=1.0),
    loss='categorical_crossentropy',
    metrics=['accuracy', Precision(name='precision'), Recall(name='recall')]
)

resnet_model.summary()
````

## üöÄ Treinamento com K-Fold

````python
# Treinamento com K-Fold:

# Armazenamento das m√©tricas por fold
val_acc_per_fold = []
val_loss_per_fold = []
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []
all_y_true = []
all_y_pred = []


kf = KFold(n_splits=5, shuffle=True, random_state=42)

for fold, (train_idx, val_idx) in enumerate(kf.split(df)):
    print(f"\nüìÅ Treinando Fold {fold + 1}...\n")

    # Split
    train_df = df.iloc[train_idx]
    val_df = df.iloc[val_idx]

    # Geradores
    train_gen = train_datagen.flow_from_dataframe(
        train_df,
        x_col='filename',
        y_col='label',
        target_size=(224, 224),
        batch_size=32,
        class_mode='categorical',
        subset='training',
        shuffle=True
    )

    val_gen = validation_datagen.flow_from_dataframe(
        val_df,
        x_col='filename',
        y_col='label',
        target_size=(224, 224),
        batch_size=32,
        shuffle=False
    )

    # Pesos das Classes (Class Weights)
    class_weights = compute_class_weight(
        class_weight='balanced',
        classes=np.unique(train_gen.classes),
        y=train_gen.classes
    )
    class_weights_dict = dict(enumerate(class_weights))

    # Diret√≥rio de log espec√≠fico por fold com timestamp
    log_dir = f"/content/drive/MyDrive/logs/fold_{fold+1}_{datetime.datetime.now().strftime('%Y%m%d-%H%M%S')}"  # se estiver trabalhando com pasta do google drive no google colab
    # log_dir = f"../outputs/logs/fold_{fold+1}_{datetime.datetime.now().strftime('%Y%m%d-%H%M%S')}"  # se estiver trabalhando com pasta local no vscode
    tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)

    # Modelo
    history = resnet_model.fit(
        train_gen,
        epochs=60,
        validation_data=val_gen,
        class_weight=class_weights_dict,
        callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5), tensorboard_callback]
    )

    with open(f'history_fold_{fold+1}.json', 'w') as f:
        json.dump(history.history, f)


    # Avalia√ß√£o final do fold
    results = resnet_model.evaluate(val_gen)
    val_acc_per_fold.append(results[1])
    val_loss_per_fold.append(results[0])

    # Salvamento no final de cada fold
    resnet_model.save(f'model_fold_{fold+1}.keras')
    print(f"\nüíæ Modelo salvo Fold {fold+1}")


    # Visualiza√ß√£o de m√©tricas - Gr√°ficos de Acur√°cia e Loss a cada fold
    best_epoch_acc = np.argmax(history.history['val_accuracy']) + 1
    best_epoch_loss = np.argmin(history.history['val_loss']) + 1

    fig, axs = plt.subplots(1, 2, figsize=(14, 5))
    fig.suptitle(f'Treinamento - Fold {fold + 1}', fontsize=16)

    # Accuracy
    axs[0].plot(history.history['accuracy'], 'r-', label='Training')
    axs[0].plot(history.history['val_accuracy'], 'g-', label='Validation')
    axs[0].scatter(best_epoch_acc, history.history['val_accuracy'][best_epoch_acc-1], color='blue', label=f'Best epoch (accuracy) = {best_epoch_acc}')
    axs[0].set_title("Training and Validation Accuracy")
    axs[0].set_xlabel("Epochs")
    axs[0].set_ylabel("Accuracy")
    axs[0].legend()

    # Loss
    axs[1].plot(history.history['loss'], 'r-', label='Training')
    axs[1].plot(history.history['val_loss'], 'g-', label='Validation')
    axs[1].scatter(best_epoch_loss, history.history['val_loss'][best_epoch_loss-1], color='blue', label=f'Best epoch (loss) = {best_epoch_loss}')
    axs[1].set_title("Training and Validation Loss")
    axs[1].set_xlabel("Epochs")
    axs[1].set_ylabel("Loss")
    axs[1].legend()

    plt.tight_layout()
    plt.show()

    # Matriz de confus√£o
    y_true = val_gen.classes
    y_pred = resnet_model.predict(val_gen)
    y_pred_classes = np.argmax(y_pred, axis=1)
    all_y_true.extend(y_true)
    all_y_pred.extend(y_pred_classes)

    labels = list(val_gen.class_indices.keys())
    cm = confusion_matrix(y_true, y_pred_classes)

    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    fig, ax = plt.subplots(figsize=(6, 6))
    disp.plot(cmap='Blues', ax=ax, values_format='d')
    plt.title("Confusion Matrix")

    acc = accuracy_score(y_true, y_pred_classes)
    prec = precision_score(y_true, y_pred_classes, average='macro')
    rec = recall_score(y_true, y_pred_classes, average='macro')
    f1 = f1_score(y_true, y_pred_classes, average='macro')

    # Salva m√©tricas do fold
    accuracy_scores.append(acc)
    precision_scores.append(prec)
    recall_scores.append(rec)
    f1_scores.append(f1)



    # Mostra m√©tricas
    plt.figtext(0.5, -0.1,
                f"Accuracy: {acc:.2f} | Precision: {prec:.2f} | Recall: {rec:.2f} | F1 Score: {f1:.2f}",
                ha='center', fontsize=12)
    plt.tight_layout()
    plt.show()


# An√°lise final das curvas de aprendizado
train_acc_all_folds, val_acc_all_folds = [], []
train_loss_all_folds, val_loss_all_folds = [], []

for fold in range(1, 6):
    with open(f'history_fold_{fold}.json', 'r') as f:
        h = json.load(f)
        train_acc_all_folds.append(h['accuracy'])
        val_acc_all_folds.append(h['val_accuracy'])
        train_loss_all_folds.append(h['loss'])
        val_loss_all_folds.append(h['val_loss'])

train_acc_all_folds = np.array(train_acc_all_folds)
val_acc_all_folds = np.array(val_acc_all_folds)
train_loss_all_folds = np.array(train_loss_all_folds)
val_loss_all_folds = np.array(val_loss_all_folds)

mean_train_acc = train_acc_all_folds.mean(axis=0)
std_train_acc = train_acc_all_folds.std(axis=0)
mean_val_acc = val_acc_all_folds.mean(axis=0)
std_val_acc = val_acc_all_folds.std(axis=0)
mean_train_loss = train_loss_all_folds.mean(axis=0)
std_train_loss = train_loss_all_folds.std(axis=0)
mean_val_loss = val_loss_all_folds.mean(axis=0)
std_val_loss = val_loss_all_folds.std(axis=0)

epochs = range(1, len(mean_val_acc) + 1)
best_epoch_acc = np.argmax(mean_val_acc) + 1
best_epoch_loss = np.argmin(mean_val_loss) + 1

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs, mean_train_acc, 'r-', label='Training Accuracy')
plt.plot(epochs, mean_val_acc, 'g-', label='Validation Accuracy')
plt.scatter(best_epoch_acc, mean_val_acc[best_epoch_acc - 1], color='blue', label=f'Best epoch: {best_epoch_acc}')
plt.fill_between(epochs, mean_val_acc - std_val_acc, mean_val_acc + std_val_acc, color='green', alpha=0.2)
plt.title('Mean Accuracy Across Folds')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs, mean_train_loss, 'r-', label='Training Loss')
plt.plot(epochs, mean_val_loss, 'g-', label='Validation Loss')
plt.scatter(best_epoch_loss, mean_val_loss[best_epoch_loss - 1], color='blue', label=f'Best epoch: {best_epoch_loss}')
plt.fill_between(epochs, mean_val_loss - std_val_loss, mean_val_loss + std_val_loss, color='green', alpha=0.2)
plt.title('Mean Loss Across Folds')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# T√≠tulo geral para os dois gr√°ficos
plt.suptitle('Resultado Final: Curvas M√©dias de Acur√°cia e Loss (K-Fold)', fontsize=16, y=1.02)

plt.tight_layout()
plt.show()


# Labels consistentes
labels = list(val_gen.class_indices.keys())

# Matriz de confus√£o final
final_cm = confusion_matrix(all_y_true, all_y_pred)

plt.figure(figsize=(8, 6))
disp = ConfusionMatrixDisplay(confusion_matrix=final_cm, display_labels=labels)
disp.plot(cmap='Blues', values_format='d')
plt.title('Matriz de Confus√£o Final - Todos os Folds')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# Resumo final dos folds
print("\nResultados Cross-Validation:")
print(f"Accuracy : {np.mean(accuracy_scores):.4f} ¬± {np.std(accuracy_scores):.4f}")
print(f"Precision: {np.mean(precision_scores):.4f} ¬± {np.std(precision_scores):.4f}")
print(f"Recall   : {np.mean(recall_scores):.4f} ¬± {np.std(recall_scores):.4f}")
print(f"F1 Score : {np.mean(f1_scores):.4f} ¬± {np.std(f1_scores):.4f}")
````

## üìã Resultados Finais da Valida√ß√£o Cruzada

````python
# Resultados Finais da Valida√ß√£o Cruzada:


# Obter nomes das classes
class_names = list(val_gen.class_indices.keys())
n_classes = len(class_names)

# Inicializar listas para m√©tricas por fold
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []
kappa_scores = []
auc_macro_scores = []
auc_scores_per_class = [[] for _ in range(n_classes)]

# Para matriz de confus√£o agregada
all_y_true = []
all_y_pred_classes = []
all_y_pred_probs = []

# Valida√ß√£o cruzada
kf = KFold(n_splits=5, shuffle=True, random_state=42)
for fold, (train_idx, val_idx) in enumerate(kf.split(df)):
    print(f"\nüîç Reavaliando Fold {fold + 1} para m√©tricas agregadas...")

    val_df = df.iloc[val_idx]

    val_gen = validation_datagen.flow_from_dataframe(
        val_df,
        x_col='filename',
        y_col='label',
        target_size=(224, 224),
        batch_size=32,
        shuffle=False
    )

    model = tf.keras.models.load_model(f'model_fold_{fold+1}.keras')

    y_pred_prob = model.predict(val_gen)
    y_pred_class = np.argmax(y_pred_prob, axis=1)
    y_true = val_gen.classes

    # Salvar para matriz de confus√£o agregada
    all_y_true.append(y_true)
    all_y_pred_classes.append(y_pred_class)
    all_y_pred_probs.append(y_pred_prob)

    # M√©tricas por fold
    acc = accuracy_score(y_true, y_pred_class)
    prec = precision_score(y_true, y_pred_class, average='macro', zero_division=0)
    rec = recall_score(y_true, y_pred_class, average='macro', zero_division=0)
    f1 = f1_score(y_true, y_pred_class, average='macro', zero_division=0)
    kappa = cohen_kappa_score(y_true, y_pred_class)

    # AUC
    y_true_bin = label_binarize(y_true, classes=range(n_classes))
    auc_class = roc_auc_score(y_true_bin, y_pred_prob, average=None, multi_class="ovr")
    auc_macro = roc_auc_score(y_true_bin, y_pred_prob, average="macro", multi_class="ovr")

    # Salvar m√©tricas
    accuracy_scores.append(acc)
    precision_scores.append(prec)
    recall_scores.append(rec)
    f1_scores.append(f1)
    kappa_scores.append(kappa)
    auc_macro_scores.append(auc_macro)
    for i in range(n_classes):
        auc_scores_per_class[i].append(auc_class[i])


# Tabela com m√©tricas gerais
metrics_summary = pd.DataFrame({
    'M√©trica': [
        'Accuracy',
        'Precision (macro)',
        'Recall (macro)',
        'F1-Score (macro)',
        'Cohen‚Äôs Kappa',
        'AUC-ROC (macro)'
    ],
    'M√©dia': [
        np.mean(accuracy_scores),
        np.mean(precision_scores),
        np.mean(recall_scores),
        np.mean(f1_scores),
        np.mean(kappa_scores),
        np.mean(auc_macro_scores)
    ],
    'Desvio Padr√£o': [
        np.std(accuracy_scores),
        np.std(precision_scores),
        np.std(recall_scores),
        np.std(f1_scores),
        np.std(kappa_scores),
        np.std(auc_macro_scores)
    ]
})


# Tabela AUC por Classe
auc_per_class_df = pd.DataFrame({
    'Classe': class_names,
    'AUC M√©dia': [np.mean(scores) for scores in auc_scores_per_class],
    'Desvio Padr√£o': [np.std(scores) for scores in auc_scores_per_class]
})


def azul_coluna_metrica(val):
    return 'background-color: #2196f3; color: white; font-weight: bold;' if isinstance(val, str) else ''

def azul_coluna_classe(val):
    return 'background-color: #2196f3; color: white; font-weight: bold;' if isinstance(val, str) else ''


# Estilo para a tabela de m√©tricas finais

# --- TABELA DE M√âTRICAS FINAIS ---

styled_metrics = metrics_summary.style\
    .format({'M√©dia': '{:.4f}', 'Desvio Padr√£o': '{:.4f}'})\
    .set_caption("M√âTRICAS FINAIS ‚Äì CROSS-VALIDATION (5 folds)")\
    .set_table_styles([
        {"selector": "caption", "props": [
            ("caption-side", "top"),
            ("font-size", "16px"),
            ("font-weight", "bold"),
            ("color", "white"),
            ("text-align", "center"),
            ("background-color", "#f35422"),
            ("border", "1px solid #ddd")
        ]},
        {"selector": "thead th", "props": [
            ("background-color", "#f35422"),
            ("color", "white"),
            ("text-align", "center"),
            ("font-size", "14px"),
            ("font-weight", "bold"),
            ("border", "1px solid #ddd")
        ]},
        {"selector": "tbody td", "props": [
            ("text-align", "center"),
            ("font-size", "13px"),
            ("border", "1px solid #ddd")
        ]}
    ], overwrite=False)\
    .apply(lambda x: ['background-color: #f2f2f2' if i % 2 == 1 else '' for i in range(len(x))], axis=0)\
    .applymap(azul_coluna_metrica, subset=['M√©trica'])


# --- TABELA DE AUC POR CLASSE ---

styled_auc = auc_per_class_df.style\
    .format({'AUC M√©dia': '{:.4f}', 'Desvio Padr√£o': '{:.4f}'})\
    .set_caption("AUC-ROC por Classe")\
    .set_table_styles([
        {"selector": "caption", "props": [
            ("caption-side", "top"),
            ("font-size", "16px"),
            ("font-weight", "bold"),
            ("color", "white"),
            ("text-align", "center"),
            ("background-color", "#f35422"),
            ("border", "1px solid #ddd")
        ]},
        {"selector": "thead th", "props": [
            ("background-color", "#f35422"),
            ("color", "white"),
            ("text-align", "center"),
            ("font-size", "14px"),
            ("font-weight", "bold"),
            ("border", "1px solid #ddd")
        ]},
        {"selector": "tbody td", "props": [
            ("text-align", "center"),
            ("font-size", "13px"),
            ("border", "1px solid #ddd")
        ]}
    ], overwrite=False)\
    .apply(lambda x: ['background-color: #f2f2f2' if i % 2 == 1 else '' for i in range(len(x))], axis=0)\
    .applymap(azul_coluna_classe, subset=['Classe'])

# Desativar visualiza√ß√£o interativa com setas
pd.options.display.html.table_schema = False

# Exibir as tabelas sem √≠ndice e sem colunas extras
display(HTML("<style>caption {caption-side: top; font-size:16px; font-weight:bold;}</style>"))
display(styled_metrics.hide(axis='index'))
display(HTML("<br><br>"))
display(styled_auc.hide(axis='index'))
````
